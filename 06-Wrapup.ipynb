#### What's New for ML in 2.3?

* Proper Image Data Type for binary/images in DataFrames: https://issues.apache.org/jira/browse/SPARK-21866
* Linear regression with Huber M-Estimation: https://issues.apache.org/jira/browse/SPARK-3181
* Improvements to GradientBoostedTree algorithms
* Ability to create custom pipeline stages from Python (*fairly limited* for now but a first step)
* FeatureHasher http://spark.apache.org/docs/latest/ml-features.html#featurehasher

#### What's New for ML in 2.4?
* Not a ton ... expect to see more big changes in the next release (Spark 3.0, sometime next year)
* Image data format https://user-images.githubusercontent.com/19235986/47330120-afad7e00-d6a9-11e8-8a0c-4340c2af727b.png
* Power Iteration Clustering already existed in the RDD ("old") API; it will be available via Spark ML (DataFrame) API
* Kolmogorov-Smirnoff Test for dataframe
* One-off fast prediction in the driver (model only, not pipeline)
* Infrastructure for model export, proof-of-concept PMML for one or two model types (not pipeline yet)
* Partial design / infrastructure for barrier mode (gang scheduling) to allow interop with MPI style frameworks
