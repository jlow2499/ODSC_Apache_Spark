Let's fix the schema. We know (from Wikimedia) that the first three columns are:
  * `project : string`
  * `page : string`
  * `requests : int`	
  * we can ignore the 4th column -- we won't be using it
  
One way to do this is to use `selectExpr` and SQL snippets to `CAST` and rename the columns:

```
wikimediaData = spark.read.option("delimiter", " ") \
                          .csv("/mnt/spark-live/pageviews.gz") \
                          .selectExpr("_c0 AS project", "_c1 AS page", "CAST(_c2 AS INT) AS requests")
```

Let's give this table (really a view, or query) a name ... so that we can look at it with SQL:

```
wikimediaData.createOrReplaceTempView("pageviews")
```

We can write all kinds of SQL queries:

* SQL:2003 + HiveQL ... documented at https://docs.databricks.com/spark/latest/spark-sql/index.html#spark-sql-language-manual) 

... and we can use a programmatic API or "domain-specific language" as well.

The DataFrame/Dataset API allows us to write native Python, Java, Scala, or R programs using Spark.

Common tasks are fairly similar across these APIs:

|SQL|DataFame API|DataFrame example (with String column names)|
|---|---|---|
|SELECT|select, selectExpr|myDataFrame.select("someColumn")|
|WHERE|filter, where|myDataFrame.filter("someColumn > 10")|
|GROUP BY|groupBy|myDataFrame.groupBy("someColumn")|
|ORDER BY|orderBy|myDataFrame.orderBy("column")|
|JOIN|join|myDataFrame.join(otherDataFrame, "innerEquiJoinColumn")|
|UNION|union|myDataFrame.union(otherDataFrame)|

The API support -- both for SQL and DataFrame/Dataset -- is far broader than these examples. E.g., columns and expressions can be written in a strongly-typed manner, using Column objects. These can be generated from Strings or Symbols; many types of JOIN are supported; and a variety of mechanisms for creating a DataFrame from a source exist in all of these APIs.

```
# Here's the earlier query with the DataFrame/Dataset API:

from pyspark.sql.functions import desc

source = spark.table("pageviews")
query = source.filter("project = 'en'").filter(source["page"].like("%Spark%")).orderBy(desc("requests"))

display(query)
```

#### Statistics APIs

Statistics (and advanced/probabilistic counting) functions live in (at least?) three places in PySpark + (at least?) one non-PySpark location

1. `pyspark.sql.DataFrameStatFunctions` (available as .stat off of any DataFrame)
  * approxQuantile
  * corr
  * cov
  * crosstab
  * freqItems
  * sampleBy (stratified sampling)
2. `org.apache.spark.sql.DataFrameStatFunctions` is the Scala source for the above, and has a couple of features not exposed to Python yet
  * bloomFilter
  * countMinSketch
3. `pyspark.sql.functions`
  * approx_count_distinct, corr
  * countDistinct, sumDistinct
  * covar_pop, covar_samp
  * kurtosis, mean, skewness
  * stddev_pop, stddev_samp
  * var_pop, var_samp
4. `pyspark.ml.stat`
  * ChiSquareTest
  * Correlation (matrix)
  * KolmogorovSmirnoffTest

#### A Note or Four on Pandas and One on `scikit-learn`

1. You can create a DataFrame from a Pandas DataFrame
  * The Pandas DF is on the driver
  * This is handy for small fact tables
  * But if your "big" data fits on the driver, then maybe you don't need/want Spark
  
2. You can convert a Spark DataFrame to a Pandas DataFrame
  * This will collect all of the data to the driver
  * This is handy for taking a sample of your data and then doing fun Python/Pandas (local) things to it
  * Or if the final stage of your processing gives you a small table
  * But if you try it with "big" data, you'll get into trouble
  
3. As of Spark 2.3, there *are* some safe and performant ways of integrating Spark and Python/SciPy/Pandas, which we'll cover at the beginning of Part 2 of this class

4. What About Koalas - https://github.com/databricks/koalas
  * What is it? A brand new wrapper for Spark DataFrame to make the API closer to that of Pandas
  * Really?
    * Get me out of here then, I already know Pandas, so I'm done!
  * Ermmm... not so fast
    * It's a new 0.1 initiative with very limited functionality, and it may be a while before it supports much of the Pandas API
    * More importantly, it hides the "impedance mismatch" between Pandas (local, in-memory) and Spark (distributed, not-in-memory, despite what you may have heard)
    * Therefore, it will make some things easier, but other things could be catastrophic (if they don't "make sense" for the backend distributed system)
  * But ... but I really like Pandas! Surely it can be made distributed in a native way?
    * Absolutely ... just not with Spark. My hat is off to Matthew Rocklin and the whole Dask team who have done exactly what you want. I love Dask!
    
5. But ... what about Scikit-learn? Is there some way to convert between Spark models and sklearn models?
  * In general, no
  * But there are 3 cool things you can do with Spark and sklearn using a library called `spark-sklearn`
    * For Linear and Logistic Regression, there is an API that will convert models between the two systems for you
    * If you have a lot of data, and some grouping key, and you want to build a separate model for each group of data, and the data for each key is small enough to train in a single machine with sklearn, you can use regular sklearn modeling/pipeline code on each group
    * If you have existing code that works with sklearn and the whole dataset is small enough to train locally, you can use Spark as a scheduler for distributed parallel grid-search hyperparam tuning
  * Learn more at https://github.com/databricks/spark-sklearn
